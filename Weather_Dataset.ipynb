IMPORTING THE LIBRARIES


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

IMPORTING THE DATASET


```python
dataset = pd.read_csv('weather_data.csv')
X= dataset.iloc[: , [0,1,3,4,5] ].values
y= dataset.iloc[: , 2].values
```

```python
print(X)
```

```python
print(y)
```

HANDLING MISSING DATA

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 2:])
X[:, 2:] = imputer.transform(X[:, 2:])
print(X)
```

ENCODING DATA BY COLUMN TRANSFER

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

SPLITTING THE DATASET INTO TRAINING AND TEST SET


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

STANDARDIZATION

```python
from sklearn.preprocessing import StandardScaler
sta = StandardScaler()
X_train[:, 11:] = sta.fit_transform(X_train[:, 11:])
X_test[:, 11:] = sta.transform(X_test[:, 11:])
```

```python
print(X_train[:, 11:])
```

```python
print(X_test[:, 11:])
```

```python
import pandas as pd

# Convert X (features) back to a DataFrame
# Note: Column names are lost after OneHotEncoder and StandardScaler. Generic names are used.
# If you need original column names, you'd need to store them before encoding/scaling.
X_df = pd.DataFrame(X)
X_df.to_csv('cleansed_features.csv', index=False)
print("Cleansed features saved to 'cleansed_features.csv'")

# Convert y (target) back to a DataFrame
y_df = pd.DataFrame(y, columns=['target_variable'])
y_df.to_csv('cleansed_target.csv', index=False)
print("Cleansed target saved to 'cleansed_target.csv'")
```
